# Introduction

The goal of this unit is to give you a wide inventory to critique experimental studies while expressing those critiques with computational models. Today, the focus is on various forms of measurement error and data contamination. Most of the time in statistical pedagogy, we assume our data obeys a series of convenient properties. Assuming our data is normally distributed is one of them. Assuming that our data was measured accurately is another. This is great for teaching because there aren't too many complications at once.

But it is bad for science. Real data always comes with wrinkles, missing values, extreme values and other hard-to-anticipate challenges. Without having a framework for dealing with this challenges, people often resort to ad hoc techniques. Cleaning the data of abnormalities is a common practice, especially by dropping bad observations. Cleaning data in this manner is sometimes fine, sometimes fraudulent, and most of the time somewhere in between. It introduces an unmodeled process that changes the data.

In the spirit we've been developing throughout the class, the better thing to do is to leave the data as is but build models that represent the processes that give rise to abnormalities. Data collection processes are also causal processes. If we can model the causal interactions between the variables we care about, we can also model what happens during data collection. Doing so will occassionally dramatically change a result. Most of the time, it won't change much. After all, a lot of data cleaning is benign. However, the advantage of modeling data abnormalities is that we can accomplish the same goals as cleaning but doing so in a principled way. 

We'll explore two examples:
- what to do with outliers?
- what to do with measurement error?

# Outliers and the Student's T distribution

In the last class, we very briefly encountered the student-t distribution. We learned it arises whenever a subgroup of our population experiences a different kind of variance than the rest of the group. We can access the student's T distribution with `stats.t()`. It has all the same methods as `stats.norm()`. It's parameters are 

    t(normality,mean,spread)
    
It ranges from 1-infinity. Here is what samples look like when `normality = 4`.

```python
import pymc as pm
import pandas as pd
import numpy as np
import arviz as az
import matplotlib.pyplot as plt
from scipy import stats
```

```python
samples = stats.t(4,0,25).rvs(2000)
plt.hist(samples,bins=50);
```

You can get a feel for how all the parameters interact by adjusting them below and plotting. Nu is the popular abbreviation for normality.

```python
nu = 4
mean = 0
std = 1
    
x = np.linspace(-5,5,500)
y = stats.t(nu,mean,std).pdf(x)
y = y / sum(y)

plt.plot(x,y)
plt.ylim([0,0.025])
```

The thing to notice is that all the visable action happens when `nu = 1,2,3`. When `n=1`, the tails become very flat, assigning a small but significant probability to observations far away from the mean. When `nu > 3`, it can be quite hard to visually see the changes in the shape of the distribution but they are there - it still does a better job of accounting for outliers. After normality gets to be about 30, the distribution is essentially normal again.

Now lets run a quick check to see whether there is likely to be hetereogeneity in our walking speeds dataset and whether that effects our estimate of the causal effect of behavioral priming. Most of the code is the same except we need one new prior and we need to swap our outcome distribution. First, let's just inspect our new model.

```python
url = 'https://raw.githubusercontent.com/daniel-saunders-phil/cogs-303/main/bargh_experiment_2_data'

data = pd.read_csv(url)

control = data[data.group == 1]
primed = data[data.group == 0]
control_speed = control.speeds.values
primed_speed = primed.speeds.values

with pm.Model() as model_1:
    m0 = pm.Normal('mean_g0',10,5)
    m1 = pm.Normal('mean_g1',10,5)
    std = pm.Gamma('std',3,1) 
    nu = pm.Exponential('nu',scale=30) + 1
    
    y0 = pm.StudentT('y0',nu=nu,mu=m0,sigma=std,observed=primed_speed)
    y1 = pm.StudentT('y1',nu=nu,mu=m1,sigma=std,observed=control_speed)
```

## Priors on the student-t distribution

How did I come up with the prior on `nu`? Our goal is to express uncertainty about whether there is heterogeneity or not. So we'd like a prior that says something like "maybe, maybe not" and puts roughly equal weight on both options. The way to realize this goal isn't very obvious. 

Part of the problem we've already run into - it's hard to put priors on variances because they are constrained to be positive. So we want to discount super high numbers like 100 and 1000 by using some kind of decaying function. That's how we arrive at the exponential shape.

Normality parameters come with an extra trick - most of the interesting stuff happens near 1. When `nu` is 1, the normal flattens out into a new shape. When `nu` is 30, the normal looks exactly the same. In between, there is a very gradual change in shape up until the numbers 3, 2, 1 when it starts to change quite rapidly. So our prior should place more weight in the region where the interesting changes are happening. The exponential has the nice property where it puts proportionally more weight on numbers that are more interesting. There are only a few numbers that change the shape a lot (1, 2, 3) and there are a lot of numbers that change the shape a little. So the exponential strikes a nice balance. This is the kind of argument that it takes a bit more formal mathematics to show rigourly. But just trust that this research has been conducted. To get a heuristic verification of the argument, however, we can look a prior predictive plot.


```python
x = np.linspace(-5,5,200)

nu_samples = pm.draw(nu,draws=20)

for n in nu_samples:
    y = stats.t(n,0,1).pdf(x)
    plt.plot(x,y,alpha=1)
```

With the prior justified, we can fit our model.

```python
with model_1:
    trace1 = pm.sample()

az.summary(trace1)
```

Notice that the mean and std didn't change. Even once we start searching for heterogeneity, they are uneffected. So that's a sign that assuming the absence of dispersion doesn't hurt much. We can draw a similar conclusion by looking at the difference in poserior distributions.

```python
diffs = trace1.posterior['mean_g0'].values.flatten() - trace1.posterior['mean_g1'].values.flatten()
plt.hist(diffs,bins=50);
```

So what did we learn? Our inference about the causal effect of priming is robust with respect to heterogeneity. You can assume heterogeneity or ignore it. Either way, it looks like priming still decreases people's walking speed by about 1 second. This is the nice thing about building multiple models. You don't have to know exactly what the causal structure is. But if you have a few plausible guesses, you can just build all the models and compare them.

## When do outlier make a difference?

The last case suggests there is little at stake in whether we use a Normal or a Student-T distribution. That is often the case. When the original data really does satisfy the assumptions of the normal distribution, then switching to a Student-T shouldn't make a difference. However, when the data contains "outliers", the two different distributions can arrive at difference conclusions.

What is an outlier, by the way? It is a bit of a philosophical puzzle. Typically, it's described as observations that are extremely unlikely according to the normal distribution. Psychology articles routinely have to deal with outliers. They often resort to dropping them individually. They want to drop them that way they can use statistical models that assume the data is normally distributed. Dropping observations is almost never justified. It creates room for ad hoc manipulation of data to generate the results we want. It violates the principle that we can learn from any data. Don't modify your data to fit your assumptions. Modify your assumptions to fit your data. 

Understanding the generative process that underlies Normal and Student-T distributions helps to clarify. When a sub-group of our observations is exposed to a different *kind* of variation, then you can get extreme values. For example, maybe one person just sprints down every hallway. They loved track and field in middle school and it became their personality in college. Or the experimenters mess up and accidentally waved down the participant before they had fully reached the elevator. These sorts of random variations are unique to subsets of individuals and are unlikely to be shared by all participants. If we can build a model of this sort of process, then we can make smarter inferences and get to use all our data.

Let's look at a case where a couple of outliers can make a difference. I'm going to strategically add three hypothetical observations to the data. I'll pick them to carefully to erase the appearance of the priming ffect. First, we'll add one sprinter to the primed group. Then I'll add two slow people to the control group - maybe the experimenter accidentally interrupted them before they reach the end of the hallway.

```python
primed_speed_outlier = np.concatenate((primed_speed,np.array([5])))
control_speed_outlier = np.concatenate((control_speed,np.array([12,11])))
```

Visually, our new observations look like this. Both look like "outliers".

```python
f,ax = plt.subplots(1,2,figsize=(10,5))

ax[0].hist(control_speed)
ax[0].hist([12,11])
ax[0].set_title("control group")
ax[1].hist(primed_speed)
ax[1].hist([5])
ax[1].set_title("primed group")
```

Okay imagine we were ignorant of this kind of heterogeneity. We just fit the old Normal distribution models. Would they still detect a priming effect?

```python
with pm.Model() as model_2:
    m0 = pm.Normal('mean_g0',10,5)
    m1 = pm.Normal('mean_g1',10,5)
    std = pm.Gamma('std',3,1) 
    
    y0 = pm.Normal('y0',m0,std,observed=primed_speed_outlier)
    y1 = pm.Normal('y1',m1,std,observed=control_speed_outlier)

    trace = pm.sample()
```

Not really. The mean effect size is near zero. There is a good chunk of the posterior that says our effect is below zero. It would be hard to argue that there was good evidence of a priming effect from this sort of graph.

```python
diffs = trace.posterior['mean_g0'].values.flatten() - trace.posterior['mean_g1'].values.flatten()
plt.hist(diffs,bins=50);
```

Now if we try to fit a student-t model, we might get different results.

```python
with pm.Model() as model_3:
    m0 = pm.Normal('mean_g0',10,5)
    m1 = pm.Normal('mean_g1',10,5)
    std = pm.Gamma('std',3,1) 
    nu = pm.Exponential('nu',scale=30) + 1
    
    y0 = pm.StudentT('y0',nu=nu,mu=m0,sigma=std,observed=primed_speed_outlier)
    y1 = pm.StudentT('y1',nu=nu,mu=m1,sigma=std,observed=control_speed_outlier)

    trace_t = pm.sample()
```

The parameter estimates are revealing. The mean of the `nu` parameter drops down to 20. That suggests the model is detecting some sub-group heterogeneity. 

Also notice that the uncertainty on this parameter is huge! The 3%-97% interval ranges from basically 1 to 66 (remember that we add 1 to the parameter to force it to stay inside it's range). The lesson is that estimating parameters like `nu` is quite difficult. You either need a lot of outliers or a lot of normally distributed data. In our case, that's okay. We don't need to know the `nu` parameter with much certainty before it starts to shift the results of `mean_g0` and `mean_g1`.

```python
az.summary(trace_t)
```

The student-t model comes to slightly different conclusions than the normal model. It thinks the effect is a bit bigger and assigns less weight to 0 and negative effects.

```python
az.style.use("arviz-doc")

diffs = trace.posterior['mean_g0'].values.flatten() - trace.posterior['mean_g1'].values.flatten()
diffs_t = trace_t.posterior['mean_g0'].values.flatten() - trace_t.posterior['mean_g1'].values.flatten()
az.plot_density([diffs,diffs_t],data_labels=['Normal','Student-T'],shade=0.2)
```

# How to represent measurement error

There are many sources of noise in an experiment. Some of them are fairly benign: the fact that some people are taller in the control group, or late for class, or sleepy. All introduce variation that we have a way of modeling with normal distributions. These sources of variation are also measured - we know how far off from the group mean each person is.

But other sources of noise aren't so nice: measurement error is one. Measurement error is a unique problem because it is unmeasured variation. All we have is the recorded speeds. We do not know how far off they are. So there might be further sources of noise then we have access to. Let's explore how to model this.

First, we'll plot all the walking speeds on the y-axis and just evenly space the participants on the x-axis.

```python
az.style.use("default")
x = np.arange(30)

plt.plot(x[:15],control_speed,'o',label="control")
plt.plot(x[15:],primed_speed,'o',label="primed")
plt.xlabel("participant")
plt.legend()
plt.ylabel("walking speed")
```

We can represent the measurement error using normal distributions. Here's the thought. The stopwatch guy will make mistakes. Sometimes he goes high. Sometimes he goes low. But mostly we should think his guess are concentrated around the true walking speed. This is an awful lot like our football field metaphor from last class. There is a baseline (being correct) and then random factors making the stopwatch man go up or down. Naturally, it's not a perfect story. The stopwatch man could be consistently low. But that's just like moving the baseline down a bit. We already know that changing the starting location doesn't change the type distribution, just the mean.

We can visualize the data points under conditions of uncertainty. We do not quite know where they are so we add random normally distributed noise to each observed measurement. Then we sample 20 partial points from our normal distribution and plot them all with high transparency. If you adjust the `std` parameter at the top, you can visualize what measurement uncertainty look like.

```python
std = 0.6
    
scatter = 20

x1 = np.repeat(np.arange(15),scatter)
y1 = stats.norm(np.repeat(control_speed,scatter),std).rvs()
x2 = np.repeat(np.arange(15,30),scatter)
y2 = stats.norm(np.repeat(primed_speed,scatter),std).rvs()

plt.plot(x1,y1,'o',alpha=0.2,color="tab:blue")
plt.plot(x2,y2,'o',alpha=0.2,color="tab:orange")
plt.ylim([6,10])
plt.xlabel("participant")
plt.ylabel("walking speed")
```

As you play with this, notice when the two groups are no longer easily distinguishable. At that point, that's enough measurement error to make learning the difference between the groups impossible. Visually, it looks like this starts to happen 0.6.

## Causal model of measurement error

Getting clear on the causal structure lurking behind our data can help us figure out what the correct model is. This is similar the example explored in lecture 17 but with some extra layers of complexity. 

```python
from graphviz import Digraph

G = Digraph()

G.node("a",'Primed')
G.node("b",'True walking speed')
G.node("c","Measured walking speed")
G.node("n","Sleepy, Tall, Late, Etc")
G.node("k","experimenter error")

G.edges(['ab'])
G.edges(['nb'])
G.edges(['bc'])
G.edges(['kc'])
G
```

Here's how we talk through the diagram. You start with an node that doesn't have arrows flowing into it. Those are nodes that don't have prior causes. In our case, let's start with `primed`. You are either in the primed group or the unprimed group. That is decided randomly by the experimenter. So we'll generate a random binary variable.

```python
sample_size = 30
primed = np.random.choice([0,1],size=sample_size)
```

We'll assume that priming actually does work. To generate a person's true walking speed, we'll need to make that a function of three things:
- some baseline typical walking speed. Here we'll assume 9 seconds.
- the priming effect. Here we'll assume being primed adds 1 second.
- Noise that arises from outside factors like being sleepy, late, or tall.

To encode those assumptions in a simulation, we write:

```python
size_of_noise = 1
baseline_walking_speed = 9
effect_of_priming = 1
noise_on_speed = np.random.normal(0,size_of_noise,size=n)

true_walking_speed =  baseline_walking_speed + effect_of_priming * primed + noise_on_speed
```

In the ordinary case, that is all we'd need. This is equivalent to the causal model at the beginning of lecture 17. But in our case, we are assuming the true walking speed doesn't quite match the observed walking speed because of measurement error. So we can take the true walking speed and distort them a little bit with an extra layer of noise. This time we'll use our empirical estimate for the amount of noise out of stopwatches.

```python
experimenter_error = np.random.normal(0,0.6489,size=n)
measured_walking_speed = true_walking_speed + experimenter_error
```

Finally, we want to divide up our data into a control group and a treatment group. We can use our `primed` variable as an index. For example, if we write:

```python
primed == 1
```

We get an array that of Booleans. We can then use that array as an index in whatever array we are interested in sorting.

```python
measured_walking_speed[primed == 1]
```

Altogether, our simulation looks like this:

```python
# simulation parameters

sample_size = 30
baseline_walking_speed = 9
effect_of_priming = 1
size_of_noise = 1

# simulation structure

primed = np.concatenate((np.zeros(sample_size // 2),np.ones(sample_size // 2))) 
                        # this is similar to generating random 0 and 1s
                        # except we can ensure we have exactly 15 of each
                        # this will simply fix an error we would run into
                        # later.                                                    

noise_on_speed = np.random.normal(0,size_of_noise,size=n)
true_walking_speed =  baseline_walking_speed + effect_of_priming * primed + noise_on_speed
experimenter_error = np.random.normal(0,0.6489,size=n)
measured_walking_speed = true_walking_speed + experimenter_error

# sorting

primed_speed_sim = measured_walking_speed[primed == 1]
control_speed_sim = measured_walking_speed[primed == 0]
```

## Bayesian models of measurement error

When programming the Bayesian model, there are two tricks. 

First, we need to distinguish between the "true" speeds and the "observed" speeds. The true speeds are governed by things that geniunely effect walking speed, priming and noise (`sig` in our model). So the first half of the model will look an awful lot like our old model. In the second half of the model, we have the observed speed. They are just the true speeds plus a little distortion based on our estimate of how much measurement error there is. If you look at the code below, you'll notice it follows the logic of the simulations written above.

```python
with pm.Model() as model_error:
    m0 = pm.Normal('mean_g0',10,5)
    m1 = pm.Normal('mean_g1',10,5)
    sig = pm.Gamma('sig',3,1)

    true_primed_speeds = pm.Normal('true_primed_speed',m0,sig,shape=15)
    true_control_speeds = pm.Normal('true_control_speed',m1,sig,shape=15)

    y0 = pm.Normal('observed_primed_speed',true_primed_speeds,0.6489,observed=primed_speed_sim)
    y1 = pm.Normal('observed_control_speed',true_control_speeds,0.6489,observed=control_speed_sim)
```

Second, we need to make a multilevel model. Each true walking speed is an unknown parameter that we'd like to estimate. We have 30 new unknowns in this case. This is why we use `shape=15` in this model - `shape=` copies the given distribution a given number of times and uses identical parameters for each copy. It helps make writing the model more compact. We could also have written:

> pm.Normal('true_primed_speed_0',m0,sig)
> pm.Normal('true_primed_speed_1',m0,sig)
> pm.Normal('true_primed_speed_2',m0,sig)
> pm.Normal('true_primed_speed_3',m0,sig)

to capture the same structure. It's more explicit but tedious and hard to change later. 

Whenenver we have an unknown in a Bayesian model, we need to put a prior on it. Having thirty parameters in a model sounds like a lot. In this case, we have more total parameters then we have observations. Surprisingly, that's okay. The reason it is okay is that we can exploit a multi-level structure. As we learn more about the average speed of each group, we can use that information to inform the priors of each individual walking speed. Meanwhile, as we learn the individual speeds, we can use that information to estimate the group means.

Maybe a metaphor will help make it clear why we use a multi-level structure. In unit 2, we had a model in which we have individual countries + a global parameter that is shared by those countries. Now we have individual people + a group parameter that is shared by those people. The two problems have identical structures. The fact that one use a globe and countries and the other uses a group and individuals is irrelevant. Any time we think of our problem as made sub-units nested inside larger units, we can use a multi-level structure. So it's an extremely flexible and useful modeling strategy. You can even imagine multi-multi-level models, with parameters at a global, national, and individual level.

The model fits, but with a little bit of trouble.

```python
with model_error:
    trace_error = pm.sample()
```

I have a single divergence. How bad could it be? Hard to say. It depends on why the model is diverging. We'll explore more about divergences later in the lecture. But whenever you have a very small number of divergences, one trick you might try is using a high `target_accept`. It ranges from 0 to 1. If you push toward 1, it causes your MCMC to run more slowly but it also makes fewer approximation errors.

```python
with model_error:
    trace_error = pm.sample(target_accept=0.95)
```

And we've cleared up the divergence. `target_accept` won't always work. For example, it wouldn't fix the problem that arises from order constraints and discontinuity. When the posterior geometry has major structural defects that make rolling a marble hard, `target_accept` won't do it. But other kinds of divergences are more benign. Here we have one of those cases.

The trace plot is pretty interesting. Notice we have 2 groups of 15 parameters for the true speeds of each observation. And we have uncertainty over the location of each true speed. The uncertainty is roughly similar to a normal distribution with 0.6489 standard deviation. This is a signal that this new model is picking up the measurement error structure.

```python
az.plot_trace(trace_error)
plt.tight_layout();
```

How about the main effect of priming? It has shrunk a little bit. We now have a *little* bit more reason to think the effect is zero. And our best guess on the size of the priming effect has decreased to around 0.75. This could likely arise just from using simulated data. Occassionally, the random number generators work out to create a small effect and sometimes they create a big effect in a sample of only 30 people.

```python
diffs = trace_error.posterior['mean_g0'].values.flatten() - trace_error.posterior['mean_g1'].values.flatten()
plt.hist(diffs,bins=50);
```

This is very surprising! I had thought that including a little measurement error would have virtually erased the effect down to zero. After all, how do you find an effect if all the data is uncertain? I was so sure that fitting a measurement error model would have blown up the effect that I said so aloud several classes ago. Here we've managed to roughly recover the effect despite multiple layers of noise. 

There is, however, a natural explanation for this result. Last lecture, we noticed that normal distributions arise whenever there is a baseline + a series of causes of variation that have an additive structure. So walking speed is increased by sleepiness, decreased by tallness and lateness. It is also increased or decreased by measurement error in the stopwatch. The stopwatch is also a source of additive variance. We noticed that if you just add bigger variations steps or more of them, you still get a normal distribution.

For example, if we add the noise caused by tallness, lateness, sleepiness to the noise caused by measurement error, we get a new normal distribution with a standard deviation around 1.19. So the total amount of noise isn't much higher. (You might wonder why the total noise isn't 1.643. Well sometimes measurement error will pull someone's speed down while being sleepy is pulling someone's speed up so noise can cancel as well as exaggerate variation.)

```python
samples = np.random.normal(0,1,size=100000) + np.random.normal(0,0.643,size=100000)
plt.hist(samples,bins=50)
np.std(samples)
```

## What could go wrong by ignoring a little measurement error?

Okay so at this point you might be wondering - why bother with this complicated measurement error model when it comes to virtually the same conclusions as the old model? That's a great question. Let's fit the old model and compare.

```python
with pm.Model() as model_ordinary:
    m0 = pm.Normal('mean_g0',10,5)
    m1 = pm.Normal('mean_g1',10,5)
    sig = pm.Gamma('sig',3,1) 
    
    y0 = pm.Normal('y0',m0,sig,observed=primed_speed_sim)
    y1 = pm.Normal('y1',m1,sig,observed=control_speed_sim)

    trace_ordinary = pm.sample()
```

Below is a `ridgeplot`. It is helpful whenever you want to systematically compare models that share some of the same parameters. Clearly, the two models agree about the main effect of interest - the means of the groups. However, they disagree about the standard deviation. The standard model has an inflated estimate of the amount of noise in the data.

```python
az.plot_forest([trace_ordinary,trace_error],kind="ridgeplot",model_names=["standard model","measurement error model"],combined=True,var_names=['mean_g0','mean_g1','sig'],figsize=(8,4))
```

This result makes good sense - the standard model doesn't have a mechanism to distinguish between measurement error and the more substantive kind of noise. So it groups them together. By itself, that doesn't bias the estimate of the main effect - the difference between the groups.

However, a common way of summarizing the effects of the study is to divide the effect size by the estimated standard deviation. It's a quantity known as **Cohen's D**:

$$D = \frac{\text{effect_size}}{\text{standard_deviation}} $$

Researchers care about this quantity because it standardizes effects to a common index. If an intervention causes a 50 second change in reaction time, is that big or small? Well it depends on whether the variance in reaction times is around 1 second or 1000 seconds.

If we overestimate the variance (as in the standard model above), it will artificially deflate cohen's D or the standardized effect size. The advantage of the measurement error model is that it helps partition the different sorts of variance in the data into the proper categories. Any time we can model the causes of variation or noise, we should. It makes all other results in the model more precise.

More generally, measurement error might not be as benign. If there is a lot of measurement error, it really can swamp the causal effects as we'll see in this final example. Similarly, measurement error can be *biased* is one group experience more rounding up while the other group experiences more rounding down. We'll explore biased measurement error in a future lecture.

## Fixing divergences with reparameterization

One final topic for the day. Let's get back to those pesky divergences. We were able to resolve just by increasing target_accept. But Sometimes we cannot. Let's trigger a case like that by tripling the amount of measurement error. I'll triple it in the simulated data.

```python
# simulation parameters

sample_size = 30
baseline_walking_speed = 9
effect_of_priming = 1
size_of_noise = 1
measurement_noise = 0.6489 * 3 # adjusted here

# simulation structure

primed = np.concatenate((np.zeros(sample_size // 2),np.ones(sample_size // 2)))                                                   
noise_on_speed = np.random.normal(0,size_of_noise,size=n)
true_walking_speed =  baseline_walking_speed + effect_of_priming * primed + noise_on_speed
experimenter_error = np.random.normal(0,measurement_noise,size=n)
measured_walking_speed = true_walking_speed + experimenter_error

# sorting

primed_speed_sim = measured_walking_speed[primed == 1]
control_speed_sim = measured_walking_speed[primed == 0]
```

And also in the bayesian model. (to exaggerate this example, I've also increased the number of samples from the MCMC and decreased the prior in the gamma distribution.)

```python
with pm.Model() as model_divergences:
    m0 = pm.Normal('mean_g0',10,5)
    m1 = pm.Normal('mean_g1',10,5)
    sig = pm.Gamma('sig',1,1)

    true_primed_speeds = pm.Normal('true_primed_speed',m0,sig,shape=15)
    true_control_speeds = pm.Normal('true_control_speed',m1,sig,shape=15)

    y0 = pm.Normal('observed_primed_speed',true_primed_speeds,3*0.6489,observed=primed_speed_sim)
    y1 = pm.Normal('observed_control_speed',true_control_speeds,3*0.6489,observed=control_speed_sim)

    trace_divergences = pm.sample(2000,target_accept=0.95)
```

`target_accept` didn't fully resolve the problem. What's worse, all the other warnings showed up too - rhat and low effective sample size! Let's investigate.

A `pair` plot shows your the location of every MCMC sample as a series of 2-dimensional slices. It takes a pair of parameters from your model and uses those parameters as a coordinate system. So a dot in these graphs tells you where the MCMC sampler visited at one point in time, from the perspective of the mean parameter and the sigma parameter. It is the best toll to understand why and where the divergences show up.

```python
az.plot_pair(trace_divergences,var_names=['mean_g0','mean_g1','sig'],divergences=True);
```

The key thing to notice is that our divergences are spatially concentrated in one region of the map. This is sign of a bad posterior geometry in that area. Some shapes are difficult to roll a marble around in on a computer. I cannot explain it better than this video can: https://www.youtube.com/watch?v=Es44-Bp1aKo. To get a full explanation of the phenomena, watch from 49:15 - 57:00.

Sometimes, there are two mathemetically equivalent models that are not computationally equivalent. That is a bizarre claim but it is true. The MCMC favourite shape to roll around in is a `normal(0,1)`. So we can break up the prior into a sequences of shapes that are easier to sample from. Here, we'll sample from a pair of `normal(0,1)`'s. But then we'll shift the mean of those samples and adjust the standard deviation with our hyper parameters.

```python
with pm.Model() as model_reparameterize:
    m0 = pm.Normal('mean_g0',10,5)
    m1 = pm.Normal('mean_g1',10,5)
    sig = pm.Gamma('sig',1,1)

    true_primed_speeds = pm.Normal('true_primed_speed',0,1,shape=15)
    true_primed_speeds = m0 + true_primed_speeds * sig
    true_control_speeds = pm.Normal('true_control_speed',0,1,shape=15)
    true_control_speeds = m1 + true_control_speeds * sig

    y0 = pm.Normal('observed_primed_speed',true_primed_speeds,3*0.6489,observed=primed_speed_sim)
    y1 = pm.Normal('observed_control_speed',true_control_speeds,3*0.6489,observed=control_speed_sim)

    trace_reparameterize = pm.sample(2000,target_accept=0.95)
```

It samples with ease. Notably, we find a lot more samples with low sigma once we reparameterize. So fixing divergences can make a genuine difference to what our MCMC algorithm reports as the posterior distribution. The old, divergence MCMC gives us a (mildly) inaccurate picture.

```python
az.plot_pair(trace_reparameterize,var_names=['mean_g0','mean_g1','sig'],divergences=True);
```

This difference is reflected in our estimates too. Mostly it is benign. But the new, reparameterized model indicates we might have very small sigma while the old model things small sigma is implausible. 

I'm inclined to agree with the reparameterized model. Because the measurement error is huge, it dominates the dataset. That means that ordinary substantial noise might be non-existent. Of course, we know it really does exist because we programmed the simulation to have it. But the posterior reflects a possibility that has some good intuitive backing - the simulation also has way more measurement error than it does substantial noise.

```python
az.plot_forest([trace_divergences,trace_reparameterize],kind="ridgeplot",model_names=["normal parameterization","alternative parameterization"],combined=True,var_names=['mean_g0','mean_g1','sig'],figsize=(8,4))
```

Reparameterization is a huge topic and a rather difficult one. I just want to introduce you to this one basic trick. It does two practical things for you:
1. You can sometimes debug a model on your own by following the same pattern.
2. If you cannot debug your model but you notice spatially clustered divergences, let me know and I can help. Maybe there is another reparameterization trick that you'll need instead.

Finally, notice that effect we've all been waiting for - modeling measurement error also erases the difference between group means. Once we have 3*0.6 measurement error, that is so much noise that we cannot tell which group has the higher mean walking speed. So the effect can disappear with an adequate model of measurement error. We just need a lot of it.

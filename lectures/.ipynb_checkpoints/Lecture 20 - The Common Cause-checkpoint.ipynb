# Direct and total effects

*reproduced from lecture 18*

Last class, we were working with a causal graph like this:

```python
dot = Digraph()

dot.node("a",'Primed')
dot.node("b",'Walking_speed')
dot.node("c","Mood")
dot.node("n","Noise on mood")
dot.node("u","Noise on walking_speed")

dot.edge('a','c')
dot.edge('a','b')
dot.edge('c','b')
dot.edge('n','c')
dot.edge('u','b')
dot
```

We explored the *direct* effect of priming on speed. That is, how causal influence goes directly from priming to speed, without travelling down any side routes. However, there is a different kind of causal estimate we might want, the *total* effect. The total effect includes every possible path from priming to mood. You estimate the total effect by just omitting every mediating variable. 

To illustrate, we'll use the same simulation.

```python
# generate all variables that are not a function of anything

sample_size = 400

primed = np.random.randint(low=0,high=2,size=sample_size) # random assignment to priming condition
noise_on_mood = stats.norm(0,2).rvs(sample_size) # noise
noise_on_walking_speed = stats.norm(0,1).rvs(sample_size) # noise

# mood = baseline mood + priming effect + noise

mood = -3 * primed + noise_on_mood

# walking speed = baseline + priming effect + mood effect + noise

walking_speed = 8 - 1 * primed + 0.5 * mood + noise_on_walking_speed
```

But this time, we'll omit the mood variable in the pymc model. Omitting the variable means the model doesn't have measurements of mood and can only explore the correlation between priming and walking speed. If the causal effect of priming travels down two different pathways, the statistical model will detect the total effect across both pathways rather than the direct effect.

```python
with pm.Model() as total_effect:
    a = pm.Normal('Intercept',10,5)
    b1 = pm.Normal('Priming',0,10)
    s = pm.Exponential('Noise',1)
    
    mean = a + b1*primed
    
    y = pm.Normal('walking_speed',mean,s,observed=walking_speed)
    
    trace = pm.sample()
    
az.plot_posterior(trace);
```

The resulting estimate is -2.5 for the priming effect. So the total effect is much larger than the direct effect. 

Why -2.5? Well -2.5 is the sum of two different causal pathways. The first causal pathway is the direct one, just -1. The second causal pathway is more complicated. But notice that priming causes a -3 effect on mood. And then mood causes a 0.5 effect on speed. And -3 * 0.5 = -1.5. The really neat part is that -1 (direct path) + -1.5 (indirect path) is -2.5 (total effect). So the causal graphs have this amazing property where the total effect is just the sum of each distinct causal pathway. It helps to put the effect coefficients directly onto the graph.

```python
dot = Digraph()

dot.node("a",'Primed')
dot.node("b",'Walking_speed')
dot.node("c","Mood")
dot.node("n","Noise on mood")
dot.node("u","Noise on walking_speed")

dot.edge('a','c','-3')
dot.edge('a','b','-1')
dot.edge('c','b','0.5')
dot.edge('n','c')
dot.edge('u','b')
dot
```

As you start exploring more complex causal models, you might start to feel unsure whether your pymc model really matches the causal structure you are after. To help you provide a way of debunking the pymc model, we have a tool that generates a graphviz picture out of any pymc model.

```python
pm.model_to_graphviz(total_effect)
```

When estimating the total effect, we are treating walking speed as a function of two things: noise and priming. So there should be two nodes causal nodes following into walking_speed. Intercept is just kind of there - it's not really interpreted as a causal effect but you need one to establish a baseline walking speed for everyone in the group.

By contrast, in the direct effect model, we treat walking speed as function of noise, priming and mood. So we need three (plus intercept) causal arrows flowing into the walking speed.

```python
with pm.Model() as direct:
    a = pm.Normal('Intercept',10,5)
    b1 = pm.Normal('Priming',0,10)
    b2 = pm.Normal('Mood',0,10)
    s = pm.Exponential('Noise',1)
    
    mean = a + b1*primed + b2 * mood
    
    y = pm.Normal('walking_speed',mean,s,observed=walking_speed)
    
```

```python
pm.model_to_graphviz(direct)
```

The lesson here is that `pm.model_to_graphviz()` is a helpful debugging tool and provides a minimal check against programming errors into the pymc models.

# The Common Cause

There are three major types of causal networks that we have to be careful around; mediators, common causes, and colliders. In each case, there is a question about which variables to include in our causal model and when to include them. Graph theory gives some answers. Today we'll explore the common cause. Next class we'll explore the collider.

Here's the motiviating story. My girlfriend gets frequent headaches. Her preferred remedy is Peppermint Halo essential oil (TM) (available at Saje for those interested). It really does seem to help a lot. Like many alternative medicine remedies, you might wonder whether they really work. We know that placebo effects can be very powerful. A placebo effect arises because people's thoughts and expectations have a large impact on the body and can often heal various illnesses, especially psychological illness. So we might develop an alternative explanation: it is really optimism about the possibility of feeling better that causes my girlfriend to feel better and it also causes her to use essential oils in the first place. If she wasn't feeling optimistic about getting better, she wouldn't bother to try the oils.

```python
dot = Digraph()
dot.graph_attr['layout'] = 'neato'

dot.node("a",'essential oils',pos="200,400")
dot.node("b",'optimism',pos="0,200")
dot.node("c","headache",pos="100,0")

dot.edge("a",'c',len='2.0')
dot.edge("b",'a',len='2.0')
dot.edge("b",'c',len='2.0')

dot
```

Our scientific goal will be to figure out how big the relative effects are: is it just optimism that works? just essential oils that work? or, most likely, a little bit of both. As usual, we build a simulation matching this graph. In this case we'll make it so optimism has a big effect but oils have a small effect.

```python
optimism = stats.norm(0,1).rvs(100)
oil = optimism
headache = 3 - 2*optimism - 0.5*oil
headache
```

## Adding noise

Our simulation is missing noise terms. Sometimes someone might try essential oils when they are not feeling optimistic or fail to try them when they are feeling optimismistc. So we need noise there. We also need noise on the headache: sometimes it gets better independent of all our treatments and sometimes it gets worse even with both treatments.

```python
dot = Digraph()
dot.graph_attr['layout'] = 'neato'

dot.node("a",'essential oils',pos="200,400")
dot.node("b",'optimism',pos="0,200")
dot.node("c","headache",pos="100,0")
dot.node('d','noise')
dot.node('e','noise')

dot.edge("a",'c',len='2.0')
dot.edge("b",'a',len='2.0')
dot.edge("b",'c',len='2.0')
dot.edge('d','c')
dot.edge('e','a',len='1.5')

dot
```

You might be surprised that I omit noise on optimism. As usual, whenever you feel something is weird in the simulation, you can play around with. Try adding noise onto the graph and then the simulation and see what happens. 

I will spoil the results of your experiment a little bit: nothing. Optimism might be a product of some complicated causal process that is invisible on this graph. But that doesn't matter to the question of the effect of optimism on headache. It doesn't matter where optimism comes from, it just matter what it does. We sorta implicitly had a noise term in our simulation: when we randomly generate the values for optimism, that's like saying they are a product of some noise process. Adding noise onto optimism is just adding noise onto noise.

There is a nice lesson here: rarely in life or science will we know the "full" causal structure. We usually only need to represent a slice of the world to get effective inferences. It's delicate around the edges, omitting mediators or common causes that are important can bias our estimates. But loads of causes can safely be left out without causing any trouble.

Anyway, here's the simulation with noise.

```python
optimism = stats.norm(0,1).rvs(100)
oil = optimism + stats.norm(0,1).rvs(100)
headache = 3 - 2*optimism - 0.5*oil + stats.norm(0,0.15).rvs(100)
```

## Overestimating the causal effect of essential oils

Now the question is, if I want to know the effect of essential oils on headache, which variables should I include? We know from our simulation that the oil effect is only -0.5. We can either include essential oils + optimism or just essential oils. Let's try it both ways and see what happens.

```python
with pm.Model() as overestimate:
    a = pm.Normal('baseline headache level',0,50)
    b = pm.Normal('oil effect',0,50)
    s = pm.Exponential('noise',1)
    
    mean = a + b*oil
    
    y = pm.Normal('headache',mean,s,observed=headache)
    
    trace = pm.sample()
    
az.plot_posterior(trace);
```

Our resulting estimate is -1.5, which is 3 times larger than it's supposed to be. If we want to know the effectiveness of essential oils, this statistical model overestimates their effectiveness by a large amount. Why? Well it's tracking correlations that are not causal correlations. There is a strong correlation between essential oils and headaches. Part of that is directly causal. But a large part of that is a correlation induced by their common cause in optimism.

Also notice that noise term is much bigger than it is supposed to be. That's a sign that our model is poorly specified. If the statistical model thinks that most of the variation in headache is just noise, that's because it cannot track the right causal network to partition the noise from the real patterns.

This estimate of -1.5 isn't very helpful. If someone was having a headache but didn't have optimism, you couldn't give them much relief just by applying the oil. They might get a little relief (-0.5) but $\frac{2}{3}$ of the correlation arises through a non-causal pathways. This is the dangerous thing about placebos: if someone already believes the medicine works, then great. But that doesn't mean the medicine will export to new contexts where people don't share those beliefs.

## Correctly estimating the effect

By contrast, putting both varibles into the model gives us great results: it learns the effectiveness of oils and optimism correctly.

```python
with pm.Model() as correct estimate:
    a = pm.Normal('baseline headache level',0,50)
    b1 = pm.Normal('oil effect',0,50)
    b2 = pm.Normal('optimism effect',0,50)
    s = pm.Exponential('noise',1)
    
    mean = a + b1*oil + b2*optimism
    
    y = pm.Normal('headache',mean,s,observed=headache)
    
    trace = pm.sample()
    
az.plot_posterior(trace);
```

There is also a bigger lesson here about why we conduct randomized controlled experiments. If you just run a study by talking to your friends about whether they use essential oils or not, you are liable to experience *confounding* from common causes. Something causes some of your friends to use essential oils and that something also helps them reduce headaches. Alternative medicines in general suffer from a problem where most of what we know about their effectiveness is testimony from our friends. There is very little corporate interest in conducting expensive randomization experiments on alternative medicines. So instead we have lots of randomized trials on pharmaceutical medicines. That might not be a good thing: alternative medicines might, in some cases, work much better than pharmaceutical medicines. The trouble is, it's hard to know. Causal graph theory explains why.

In an experiment, we could control who takes essential oils and who doesn't. The experimenter decides who is in what group, not any other variable. So randomization is like erasing a causal arrow that flows into the treatment you are studying.

```python
dot = Digraph()
dot.graph_attr['layout'] = 'neato'

dot.node("a",'essential oils',pos="200,400")
dot.node("b",'optimism',pos="0,200")
dot.node("c","headache",pos="100,0")
dot.node('d','noise')
dot.node('e','noise')

dot.edge("a",'c',len='2.0')
dot.edge("b",'c',len='2.0')
dot.edge('d','c')
dot.edge('e','a',len='1.5')

dot
```

But we learned that this causal graph can really just be simplified into:

```python
dot = Digraph()

dot.node("a",'essential oils')
dot.node("c","headache")
dot.node('e','noise')

dot.edge("a",'c',len='2.0')
dot.edge('e','c',len='1.5')

dot
```

# David Hume Strikes Again

David Hume is an annoying guy. His most famous contribution to philosophy is the problem of induction, a deeply wicked problem. The problem of induction has this odd character where it looks really stupid at first ("how do you know the sun will rise tomorrow") but the more you think about it the more profound it becomes. Hume's second most famous contribution to philosophy is just as annoying. How do you know what causes what?

The motivation for the problem of causation starts offs with a simple observation: we never see causation happening. Causation is an invisible force that operates underneath what we observe. We see data. We see correlations in our data. But correlations and data do not provide sufficient information to tell us what causes what.

For example, suppose you noticed that two variables $A, B$ are tightly correlated: whenever one is big, the other is big. Whenever one is small, the other is small. Based on that information, how could you tell whether $A$ causes $B$ or the other way around?

```python
dot = Digraph()
dot.node('a')
dot.node('b')

dot.node('c','a')
dot.node('d','b')

dot.edge('d','c')
dot.edge('a','b')
dot
```

You might be tempted to think the answer is: time. If A happens first, then it causes B. If B happens first, then it causes A. That's a helpful thought. But it won't work in all cases. There are plenty of things that happen first but are not causes. There are other things that happen second but are not effect. In more complex causal graphs, this is a realistic possibility. Suppose essential oils do not work. Optimism is what alleviates headaches. Someone might still regularly apply oils before they get headache relief. So temporal order doesn't offer a sure path to causal inference. We can make the problem a bit more general by exploring bigger causal graphs. How would you ever tell the difference between these two causal graphs?

```python
dot = Digraph()

dot.graph_attr['layout'] = 'neato'

dot.node("d",'b',pos="200,400")
dot.node("e",'a',pos="0,200")
dot.node("f",'c',pos="100,0")

dot.edge("e",'d',len='1.0')
dot.edge("e",'f',len='1.0')
dot.edge("d",'f',len='1.0')

dot.node("a")
dot.node("b")
dot.node("c")

dot.edges(['ac'])
dot.edges(['ab'])
dot.edges(['cb'])
dot
```

The only difference is whether $b$ causes $c$ or whether $c$ causes $b$. You can build a simulation where the data look identical in both cases. The correlations look identical as well. And temporal order might not be helpful because of the common cause found in $a$. The point generalizes. For any given pattern of data, there are multiple causal structures that might be responsible for the data. The data alone cannot tell you which causal graph you are in.

If you feel comfortable with this point, it's easier to see where Hume goes next. If science is about what is objective, then that must be visible to everyone and something we can all agree on. The observations, the data, for example, are the objective bits that we can all agree on. This puts causation in a weird place: is causation subjective? does it belong in science? Hume raised this questions and no one really had a good answer at the time. In the Pearl reading, he picks up the story about 150 years later with Karl Pearson. Pearson, an early, influential statistician, followed this line of thought to its logical conclusion. Science can only study what is objective. That amounts to the data and the correlations in the data, in Pearson's eyes. Causation as a concept should be abolished. Pearl does a nice job of exploring the history in the early 1900s.

## The counterfactual interpretation of causation

The "solution" to Hume's puzzle about causation follows a similar pattern to his puzzle about induction. With induction, we needed to get a little bit more modest about what we wanted from inferential reasoning, moving from true generalizations to high posterior probabilities. We also needed to add prior assumptions. In the causation case, we have to get a little bit more specific about what we mean by "causation", the semantics. And we need to start making some assumptions, the epistemology.

The modern interpretation of causation taken up by causal graph theory is a *counterfactual* one. When we say "$A$ causes $B$", we mean that "If I changed $A$ then I would change $B$." It's called the counter-factual interpretation because it requires you to imagine the a world where the facts are slightly different than they really are. In this case, I imagine what would happen to $B$ if $A$ were different. We can make this concrete: when I say "the cue ball caused the 8 ball to sink into the pocket", what I mean is something like "If I hadn't struck the cue ball, the 8 ball wouldn't have moved. But if I do strike the cue ball, then the 8 ball will sink into the pocket."

Counterfactual relationships are not like statistical correlations. As we saw earlier, two things can be correlated but that doesn't tell you much about whether one causes the other. Counterfactual relationships give you information on how you can change one thing to change another. These relationships are usually asymmetrical - you can move the cue ball to cause the 8 ball to move. But if you just picked up the 8 ball and slid it into the pocket, the cue ball won't suddenly move to the center of the table.

## Causal learning

Now that we are bit more clear on what causal claims mean, how can you learn them? Like in the induction case, Hume's surprising pessimistic story is mostly right. Fundamentally, we still have to guess about causal relationships. There is no simple rule to tell us what causes what just like there is no simple rule for inferring the right inductive hypothesis. But causal graph theory and statistics provide a principled way of making those guesses. We move in two steps: learning about general causal structures through experience and theory and then estimating the strength of causal effects through Bayesian inference.

The most fundamental way to learn about causation is to move about your environment and manipulate things. Perform experiments. Strike billard balls. Take drugs. Take essential oils. See what happens. These can help supply us with plausible hypotheses about what nodes and arrows to draw onto a causal graph. Sometimes you don't have direct access to a causal system. Instead you have to speculate from a far. But having well-developed scientific theories can help you transport your knowledge from one system to another. That's how we typically learn about physics - we study simple toy machines, bouncing balls and rubberbands. Then we export what we learn on the simple cases to building bridges and rocket ships. These theories help us fish out what is a plausible network of causal interactions.

So theory and interactions can help us set up the causal graph. But they aren't very good at giving us a precise idea of how big or small causal effects are. That is where statistical models come in. Once we have the graph in mind, then we just need to apply the rules of causal graph theory to tell us which variables need to be measured and included in our causal graph. In this way, causal graph theory provides a bridge between our theories and our statistical models. It tells us, exactly, what our statistical model should look for any given theory about causal interactions.

The success of a statistical estimate depends on getting the causal graph right. If, for example, you didn't know about the optimism variable in the essential oils example (or you knew about it but couldn't measure it), then you would overestimate the effectiveness of optimism.
